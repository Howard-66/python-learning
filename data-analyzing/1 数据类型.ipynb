## Numbers 数字
> 以下将逐一介绍常用的3种数字类型

```python
# 将整数赋值给变量
a = 20

# 同时将多个数值依次赋值给多个变量
b, c = 5.52, True

# 在python当中，你可以直接键入比较的表达式
d = 12 > 21

# 在屏幕上显示变量的数值
print(a, b, c, d)

# 在屏幕中显示变量的类型
print(type(a), type(b), type(c), type(d))
```

## string 字符串

>String 字符串，**即Excel当中的文本类型，一般身份证、ID、电话都会采用该格式存储**

```python
# 在python当中，需要使用 '' 单引号 或者 "" 双引号 或者 """ """ 三引号 将内容框起来，即可得到字符串
s1 = '欢迎使用python'
s2 = "Hello World"
s3 = """\n数据分析常用的平台工具包括：\n1.Excel\n2.BI\n3.R/SPSS\n4.Python\n5...."""

print(s1)
print(s2)
print(s3)
```

### ✨dict

>Dictionary 字典，本课程的最重要一种数据类型之一。**它的结构，是一种映射结构。**在Excel中，可以表现为，某一完整的列、多列；某一行、多行

- 字典在python当中的写法为：  

$$
\{key: value\}
$$ 

||在Excel当中，字典表示某列👇||
|:---|---:|:---:|

```python
{'content': 
    ['这熟悉的bgm让人灵魂颤抖，向敌军进攻，冲啊！',
    'DNA动了',
    '大制作啊',
    '好家伙 我直接好家伙',
    '大制作',
    '打卡',
    '好家伙  还有选关呢',
    '人类高质量片头',
    '感觉跟看电影一样哈哈哈哈',
    '好家伙，之前零接触的人听了整节课，真的很烧脑。']}
```

```python
{'a': 123, 'b': 456, 'c': 777}
```

- 当我们用字典表示一列的时候：
        只有一个键值对
        key是字段名，类型为字符串
        value为列内容，类型为列表  

- 每一个key: value 我们称它为一个键(key)值(value)对，如下公式，有3个键值对，他们之间用逗号彼此分开  

$$
\{key_1:value_1, key_2:value_2, key_3:value_3\}
$$ 

||在Excel当中，字典表示某行👇||
|:---|---:|:---:|

```python
{'id': '53766889040208384',
 'progress': 2691,
 'mode': 1,
 'fontsize': 25,
 'color': 16777215,
 'midHash': 'f24b7681',
 'content': '这熟悉的bgm让人灵魂颤抖，向敌军进攻，冲啊！',
 'ctime': 1629643402,
 'weight': 4,
 'idStr': '53766889040208384',
 'seg_index': 1,
 '视频内时间': 2.691,
 '弹幕发送时间': '2021-08-22 22:43:22'}
```

- 当我们用字典表示某一行的时候：
        有多个键值对
        有多份key，分别是每一个字段，类型都是字符串；
        有多个value，分别是key在该行对应的值，类型是各自的类型

- 🕵️dict也是本课程的最重要知识点之一，更多的相关内容不会一次性在一个小节内呈现。下方会不断的涉及dict的讲解

### ✨dictの查询

前面我们有提到，dict是一种映射结构，如何去理解映射结构？  
- 其中，key与value为映射的关系。你可以简单理解为，我们通过输入key，程序即会给我们返回它所对应的value  
- 例如，你打开这份Excel表格，想要查看用户都发了什么弹幕。**你第一步应该会去找《content》这个字段名在哪（key），Excel找到后，此时你便能看到具体内容（value）**

当你CTRL + F，输入content后，Excel会给你定位到相应的位置，这时候弹幕的内容也映入你的眼帘  
接下来让我们对比以下Python是如何进行这个操作的  

还是使用经典的例子

想要得到弹幕的内容，我们同样需要输入具体的字段的名字

```python
{'content': 
    ['这熟悉的bgm让人灵魂颤抖，向敌军进攻，冲啊！',
    'DNA动了',
    '大制作啊',
    '好家伙 我直接好家伙',
    '大制作',
    '打卡',
    '好家伙  还有选关呢',
    '人类高质量片头',
    '感觉跟看电影一样哈哈哈哈',
    '好家伙，之前零接触的人听了整节课，真的很烧脑。']}['content']
```

通过方括号`[]` + 具体字段名 的方式，能够得到字段对应的内容，同时，让我们用变量给这个字典赋一个别名，会更加方便我们的操作
